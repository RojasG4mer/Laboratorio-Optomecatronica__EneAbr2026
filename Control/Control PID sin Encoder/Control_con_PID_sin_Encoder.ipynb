```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P = pd.read_csv('PID sin encoder - P.csv')
PI = pd.read_csv('PID sin encoder - PI.csv')
PID = pd.read_csv('PID sin encoder - PID.csv')
columnasP = P.columns.tolist()
columnasPI = PI.columns.tolist()
columnasPID = PID.columns.tolist()
print(columnasP)
print(columnasPI)
print(columnasPID)
```

## Grafica de P

```python
plt.figure(figsize=(10, 6))
plt.plot(P[columnasP[3]], P[columnasP[4]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.title('Control P sin Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PI

```python
plt.figure(figsize=(10, 6))
plt.plot(PI[columnasP[3]], PI[columnasP[4]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.title('Control PI sin Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PID

```python
plt.figure(figsize=(10, 6))
plt.plot(PID[columnasPID[2]] * 0.2, PID[columnasPID[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.title('Control PID sin Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Las 3 juntas

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(P[columnasP[3]], P[columnasP[4]], label='Posición', color='blue')
axs[0].set_xlabel('Tiempo (s)')
axs[0].set_ylabel('Amplitud (V)')
axs[0].set_title('Control P sin Encoder')
axs[0].legend()
axs[0].grid()

axs[1].plot(PI[columnasP[3]], PI[columnasP[4]], label='Posición', color='blue')
axs[1].set_xlabel('Tiempo (s)')
axs[1].set_ylabel('Amplitud (V)')
axs[1].set_title('Control PI sin Encoder')
axs[1].legend()
axs[1].grid()

axs[2].plot(PID[columnasPID[2]] * 0.2, PID[columnasPID[3]], label='Posición', color='blue')
axs[2].set_xlabel('Tiempo (s)')
axs[2].set_ylabel('Amplitud (V)')
axs[2].set_title('Control PID sin Encoder')
axs[2].legend()
axs[2].grid()

plt.show()
```
