# Graficas de control sin PID

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sin = r'sin_histeresis.csv'
con = r'histeresis.csv'
sin_histeresis = pd.read_csv(sin)
histeresis = pd.read_csv(con)
```

```python
cols_sin = sin_histeresis.columns
cols_con = histeresis.columns
```

```python
print((cols_con))
print(len(cols_con)/6)
```

```python
DC_sinH = [90, 80, 70, 60, 50, 40, 30, 20, 10]
for i in range(9):
    print(np.abs(sin_histeresis[cols_sin[6 * i+5]]).min(),' ---  ',DC_sinH[i])
```

```python
for i in range(17):
    print(cols_con[6 * i+2],' ---  ',cols_con[6 * i + 3])
```

## Sin histeresis

```python
plt.figure(figsize=(10, 5))

for i in range(9):
    plt.plot(DC_sinH[i] , np.abs(sin_histeresis[cols_sin[6 * i+5]]).max(), marker = 'o', color = 'red')#, label=cols_sin[6 * i+2])
    
plt.grid()
plt.title('Caracterizacion de control sin ventana de histeresis')
plt.xlabel('DC (%)')
plt.ylabel('Amplitud de Error máximo (V)')
plt.show()
```

## Con histeresis

Repetimos el experimento a 3 diferentes DC y con diferentes ventanas, debemos graficar la ventana contra el error a cada DC 

```python
print((cols_con))
print(len(cols_con)/6)
```

```python
Ventanas_75 = [1, 5, 5.5, 4, 4.65]
Ventanas_50 = [4.65, 4.35, 4, 5, 5.5, 5.35]
Ventanas_25 = [5.35, 5, 5.25, 5.15, 5.1, 5.05]
```

```python
for i in range(len(Ventanas_50)):
    print(cols_con[6 * i+2],' ---  ',cols_con[6 * (i + 5 ) + len(Ventanas_75) ])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
# Ventana vs Error promedio a DC 75%
for i in range(len(Ventanas_75)):
    axs[0].plot(Ventanas_75[i], np.abs(histeresis[cols_con[6 * i+5]]).mean(), marker='o', label=Ventanas_75[i])
axs[0].set_title('Ventana de Histeresis vs Error Promedio a DC 75%')
axs[0].set_xlabel('Ventana de Histeresis (V)')
axs[0].set_ylabel('Error Promedio (V)')
axs[0].grid()
# axs[0].legend()

# Ventana vs Error promedio a DC 50%
for i in range(len(Ventanas_50)):
    axs[1].plot(Ventanas_50[i], np.abs(histeresis[cols_con[6 * (i + 5 ) + len(Ventanas_75) ]]).mean(), marker='o', label=Ventanas_50[i])
axs[1].set_title('Ventana de Histeresis vs Error Promedio a DC 50%')
axs[1].set_xlabel('Ventana de Histeresis (V)')
axs[1].set_ylabel('Error Promedio (V)')
axs[1].grid()
# axs[1].legend()

# Ventana vs Error promedio a DC 25%
for i in range(len(Ventanas_25)):
    axs[2].plot(Ventanas_25[i], np.abs(histeresis[cols_con[6 * (i + 5 ) + len(Ventanas_75) + len(Ventanas_50) ]]).mean(), marker='o', label=Ventanas_25[i])
axs[2].set_title('Ventana de Histeresis vs Error Promedio a DC 25%')
axs[2].set_xlabel('Ventana de Histeresis (V)')
axs[2].set_ylabel('Error Promedio (V)')
axs[2].grid()
# axs[2].legend()


plt.show()
```
