# Graficas de control sin PID

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sin = r'sin_histeresis.csv'
con = r'histeresis.csv'
sin_histeresis = pd.read_csv(sin)
histeresis = pd.read_csv(con)
```

```python
cols_sin = sin_histeresis.columns
cols_con = histeresis.columns
```

```python
print((cols_con))
print(len(cols_con)/6)
```

```python
DC_sinH = [90, 80, 70, 60, 50, 40, 30, 20, 10]
for i in range(9):
    print(np.abs(sin_histeresis[cols_sin[6 * i+5]]).min(),' ---  ',DC_sinH[i])
```

```python
for i in range(17):
    print(cols_con[6 * i+2],' ---  ',cols_con[6 * i + 3])
```

## Sin histeresis

```python
plt.figure(figsize=(10, 5))

for i in range(9):
    plt.plot(DC_sinH[i] , np.abs(sin_histeresis[cols_sin[6 * i+5]]).min(), marker = 'o', color = 'red')#, label=cols_sin[6 * i+2])
    
plt.grid()
plt.title('Caracterizacion de control sin ventana de histeresis')
plt.xlabel('DC (%)')
plt.ylabel('Amplitud de Error (V)')
plt.show()
```

## Con histeresis

```python
plt.figure(figsize=(20, 10))

for i in range(17):
    plt.plot(histeresis[cols_con[2]] * 0.2 , histeresis[cols_con[6 * i + 3]], label=cols_con[6 * i+2])
plt.legend()
plt.grid()
plt.title('Caracterizacion de control con ventana de histeresis')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.show()
```
