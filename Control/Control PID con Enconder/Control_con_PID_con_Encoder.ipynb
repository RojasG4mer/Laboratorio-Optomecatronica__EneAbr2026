```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P = pd.read_csv('PID con encoder - P.csv')
PI = pd.read_csv('PID con encoder - PI.csv')
PID = pd.read_csv('PID con encoder - PID.csv')
columnasP = P.columns.tolist()
columnasPI = PI.columns.tolist()
columnasPID = PID.columns.tolist()
print(columnasP)
print(columnasPI)
print(columnasPID)
```

```python
PID_corregido = pd.read_csv('PID con encoder - Ejercicio 4.csv')
columnasPID_correccion = PID_corregido.columns.tolist()
print(columnasPID_correccion)
```

## Grafica de P

```python
plt.figure(figsize=(10, 6))
plt.plot(P[columnasP[2]] * 0.3, P[columnasP[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control P con Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PI

```python
plt.figure(figsize=(10, 6))
plt.plot(PI[columnasP[2]] * 0.2, PI[columnasP[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control PI con Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PID

```python
plt.figure(figsize=(10, 6))
plt.plot(PID[columnasPID[2]] * 0.2, PID[columnasPID[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control PID con Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Las 3 juntas

```python
n = 60000 # a minutos desde milisegundos
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
axs[0].plot(P[columnasP[2]] / n, P[columnasP[3]], label='Posición', color='blue')
axs[0].plot(P[columnasP[2]] / n, P[columnasP[1]], label='Setpoint', color='red')
axs[0].set_xlabel('Tiempo (min)')
axs[0].set_ylabel('Amplitud')
axs[0].set_title('Control P con Encoder')
axs[0].legend()
axs[0].grid()

axs[1].plot(PI[columnasP[2]] / n, PI[columnasP[3]], label='Posición', color='blue')
axs[1].plot(PI[columnasP[2]] / n, PI[columnasP[1]], label='Setpoint', color='red')
axs[1].set_xlabel('Tiempo (min)')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Control PI con Encoder')
axs[1].legend()
axs[1].grid()

axs[2].plot(PID[columnasPID[2]] / n, PID[columnasPID[3]], label='Posición', color='blue')
axs[2].plot(PID[columnasPID[2]] / n, PID[columnasPID[1]], label='Setpoint', color='red')
axs[2].set_xlabel('Tiempo (min)')
axs[2].set_ylabel('Amplitud')
axs[2].set_title('Control PID con Encoder')
axs[2].legend()
axs[2].grid()

plt.show()
```
