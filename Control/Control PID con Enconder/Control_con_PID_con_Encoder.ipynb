```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P = pd.read_csv('PID con encoder - P.csv')
PI = pd.read_csv('PID con encoder - PI.csv')
PID = pd.read_csv('PID con encoder - PID.csv')
columnasP = P.columns.tolist()
columnasPI = PI.columns.tolist()
columnasPID = PID.columns.tolist()
print(columnasP)
print(columnasPI)
print(columnasPID)
```

## Grafica de P

```python
plt.figure(figsize=(10, 6))
plt.plot(P[columnasP[2]] * 0.3, P[columnasP[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control P con Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PI

```python
plt.figure(figsize=(10, 6))
plt.plot(PI[columnasP[2]] * 0.2, PI[columnasP[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control PI con Encoder')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PID

```python
plt.figure(figsize=(10, 6))
plt.plot(PID[columnasPID[2]] * 0.2, PID[columnasPID[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control PID con Encoder')
plt.legend()
plt.grid()
plt.show()
```
