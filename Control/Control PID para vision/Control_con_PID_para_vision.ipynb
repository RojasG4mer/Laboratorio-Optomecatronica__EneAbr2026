```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P = pd.read_csv('PID para camaras - P.csv')
PI = pd.read_csv('PID para camaras - PI.csv')
PID = pd.read_csv('PID para camaras - PID.csv')
columnasP = P.columns.tolist()
columnasPI = PI.columns.tolist()
columnasPID = PID.columns.tolist()
print(columnasP)
print(columnasPI)
print(columnasPID)
```

## Grafica de P

```python
plt.figure(figsize=(10, 6))
plt.plot(P[columnasP[2]] * 0.2, P[columnasP[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control P para Visión')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PI

```python
plt.figure(figsize=(10, 6))
plt.plot(PI[columnasP[2]] * 0.2, PI[columnasP[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control PI para Visión')
plt.legend()
plt.grid()
plt.show()
```

## Grafica para PID

```python
plt.figure(figsize=(10, 6))
plt.plot(PID[columnasPID[2]] * 0.2, PID[columnasPID[3]], label='Posición', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Control PID para Visión')
plt.legend()
plt.grid()
plt.show()
```

## Las 3 juntas

```python
n = 60000 # a minutos desde milisegundos
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].plot(np.round(P[columnasP[2]] * (1/n), 5), P[columnasP[3]], label = 'PV', color='blue')
axs[0].plot(np.round(P[columnasP[2]] * (1/n), 5), P[columnasP[1]], label = 'Setpoint', color='red')
axs[0].set_xlabel('Tiempo (min)')
axs[0].set_ylabel('Amplitud')
axs[0].set_title('Control P para Visión')
axs[0].grid()

axs[1].plot(np.round(PI[columnasP[2]]* (1/n), 5), PI[columnasP[3]], label = 'PV', color='green')
axs[1].plot(np.round(PI[columnasP[2]]* (1/n), 5), PI[columnasP[1]], label = 'Setpoint', color='red')
axs[1].set_xlabel('Tiempo (min)')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Control PI para Visión')
axs[1].grid()

axs[2].plot(np.round(PID[columnasPID[2]] * (1/n), 5), PID[columnasPID[3]], label = 'PV', color='orange')
axs[2].plot(np.round(PID[columnasPID[2]] * (1/n), 5), PID[columnasPID[1]], label = 'Setpoint', color='red')
axs[2].set_xlabel('Tiempo (min)')
axs[2].set_ylabel('Amplitud')
axs[2].set_title('Control PID para Visión')
axs[2].grid()

plt.show()
```
